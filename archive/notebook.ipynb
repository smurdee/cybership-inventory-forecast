```python
# params
item_id = "ATC-HPA-CRBN-128-MBLACK"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('daily_quantities.csv')
df.describe()
```

```python
df.info()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

```python
# filter for item
df = df[df['item_id'] == item_id]
df = df[df['demand'] > 0]

# plot linechart, one line per item_id
plt.plot(df['timestamp'], df['demand'], label=f'{item_id}')
plt.legend()
plt.show()
```
